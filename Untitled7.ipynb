```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams["axes.grid"] = True
plt.rcParams["font.size"] = 11
df.head()
```

```python
utm_cols = ["utm_source", "utm_medium", "utm_campaign", "utm_referrer"]
def get_utm(url, key):
    if pd.isna(url) or "?" not in url:
        return None

    params = url.split("?")[1].split("&")
    for param in params:
        if param.startswith(key + "="):
            return param.split("=")[1]
    return None
for utm in utm_cols:
    df[utm] = (
        df[utm]
        .fillna(df["utm_referrer"].apply(lambda x: get_utm(x, utm)))
        .str.lower()
        .str.strip()
    )
df[utm_cols] = df[utm_cols].fillna("no_utm")
columns = [
    "Date Created",
    "Lead stage",
    "COURSE",
    "utm_source",
    "utm_medium",
    "utm_campaign",
    "utm_referrer"
]

df = df[columns]
df["Date Created"] = pd.to_datetime(df["Date Created"], errors="coerce")
df = df.dropna(subset=["Date Created", "Lead stage"])
df
```

Были оставлены только ключевые поля, необходимые для анализа.
Дата создания лида приведена к формату datetime.
Некорректные и пустые строки удалены.

```python
leads_by_date = (
    df
    .groupby(df["Date Created"].dt.date)
    .size()
)

leads_by_date
```

пик активности,который может быть связаны с запуском рекламных кампаний.

```python
utm_cols = [
    "utm_source",
    "utm_medium",
    "utm_campaign",
    "utm_referrer"
]

df[utm_cols] = df[utm_cols].fillna("unknown")

for col in utm_cols:
    df[col] = df[col].str.lower().str.strip()
print(df[utm_cols].isna().sum())

print(df[utm_cols].head(10))
```

UTM-метки были нормализованы для корректной агрегации
и устранения дубликатов.


```python
df["is_success"] = df["Lead stage"].str.contains("получил материалы", case=False)

conversion = (
    df
    .groupby("utm_source")["is_success"]
    .mean()
    .sort_values(ascending=False)
    * 100
)
conversion
```

```python
conversion.plot(
    kind="bar",
    title="Конверсия по источникам (%)"
)

plt.ylabel("Конверсия, %")
plt.show()
```

В качестве целевого действия был выбран статус
«Получил материалы», который интерпретируется
как успешный лид.


```python
campaign_analysis = (
    df
    .groupby("utm_campaign")
    .agg(
        leads_count=("utm_campaign", "count"),
        conversion_rate=("is_success", "mean")
    )
    .reset_index()
)

campaign_analysis["conversion_rate"] *= 100

campaign_analysis
```

```python
campaign_analysis = campaign_analysis[
    campaign_analysis["leads_count"] >= 3
]

campaign_analysis
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(
    campaign_analysis["leads_count"],
    campaign_analysis["conversion_rate"]
)

for _, row in campaign_analysis.iterrows():
    plt.text(
        row["leads_count"] + 0.1,
        row["conversion_rate"] + 0.1,
        row["utm_campaign"],
        fontsize=8
    )

plt.title("Эффективность рекламных кампаний")
plt.xlabel("Количество лидов")
plt.ylabel("Конверсия (%)")
plt.grid(True)
plt.show()
```

рез-ты кампании: пик активности без идентифицированных источников.

Итоговый вывод по анализу лидов
