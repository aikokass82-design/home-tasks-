```python
import pandas as pd
import sqlalchemy
from google.colab import drive
drive.mount('/content/drive')
engine = sqlalchemy.create_engine("sqlite+pysqlite:////content/drive/MyDrive/Colab Notebooks/orders_and_users.db", echo=False)
```

```python
query = sqlalchemy.text("""
WITH customers_3 AS(
   SELECT customers.CustomerID, COUNT (orders.CustomerID)
   FROM customers
   LEFT JOIN orders
   ON orders.CustomerID = customers.CustomerID
   GROUP BY customers.CustomerID
   ORDER BY COUNT (orders.CustomerID) DESC
   LIMIT 3
   OFFSET 4
   ), part_orders AS (
   SELECT orders.CustomerID,orders.InvoiceNo,orders.Description,orders.Quantity,orders.UnitPrice,orders.Category,orders.Discount,Quantity * UnitPrice * (1 - Discount) AS total_sum
   FROM orders, customers_3
   WHERE orders.CustomerID = customers_3.CustomerID
)
SELECT *
FROM part_orders
""")
table_part = pd.read_sql(query, con=engine)
table_part
```

```python
query = ("""
WITH user_totals AS (
    SELECT
        customers.CustomerID,
        SUM(orders.Quantity * orders.UnitPrice * (1 - orders.Discount)) AS total_amount
    FROM customers
    LEFT JOIN orders
        ON customers.CustomerID = orders.CustomerID
    GROUP BY customers.CustomerID
)

SELECT
    CustomerID,
    total_amount,
    CASE
        WHEN total_amount < 500 THEN 'Новичок'
        WHEN total_amount < 1000 THEN 'Постоянный покупатель'
        WHEN total_amount < 1500 THEN 'Продвинутый клиент'
        ELSE 'VIP-клиент'
    END AS customer_group
FROM user_totals
ORDER BY total_amount DESC;""")
result_sql = pd.read_sql(query, con=engine)
result_sql
```

```python
query = (""" WITH users_total AS(
    SELECT
        CustomerID,
        SUM(orders.Quantity * orders.UnitPrice * (1 - orders.Discount)) AS total_amount
    FROM orders
    GROUP BY CustomerID
),
top_user AS (
  SELECT CustomerID
  FROM users_total
  ORDER BY total_amount DESC
  LIMIT 1
),
user_orders AS (
  SELECT
  CustomerID,
  InvoiceNo,
  Quantity * UnitPrice * (1 - Discount) AS total_sum
  FROM orders
  WHERE CustomerID IN (SELECT CustomerID FROM top_user)
)
SELECT
  CustomerID,
  InvoiceNo,
  total_sum,
  SUM(total_sum) OVER (
  ORDER BY total_sum ASC  -- от самого дешевого к самому дорогому
) AS running_total
FROM user_orders
ORDER BY total_sum ASC;
""")
result_sql = pd.read_sql(query, con=engine)
result_sql
```

```python
query = ("""WITH product_totals AS (
    -- Считаем сумму каждого товара в каждой категории
    SELECT
        Category,
        Description,
        SUM(Quantity * UnitPrice * (1 - Discount)) AS total_sum
    FROM orders
    GROUP BY Category, Description
)

SELECT
    Category,
    Description,
    total_sum,
    SUM(total_sum) OVER (
        PARTITION BY Category
        ORDER BY total_sum ASC  -- кумулятивная сумма по товарам в категории от меньшего к большему
    ) AS running_total
FROM product_totals
ORDER BY Category, total_sum ASC;
""")
result_sql = pd.read_sql(query, con=engine)
result_sql
```

```python
query = ("""WITH user_totals AS (
    SELECT
        CustomerID,
        SUM(Quantity * UnitPrice * (1 - Discount)) AS total_amount
    FROM orders
    GROUP BY CustomerID
)

SELECT
    CustomerID,
    total_amount,
    RANK() OVER (ORDER BY total_amount DESC) AS rank
FROM user_totals
ORDER BY rank;
""")
result_sql = pd.read_sql(query, con=engine)
result_sql
```
