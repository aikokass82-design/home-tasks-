```python
import pandas as pd
import sqlalchemy
from google.colab import drive
drive.mount('/content/drive')
```

```python
engine = sqlalchemy.create_engine("sqlite+pysqlite:////content/drive/MyDrive/Colab Notebooks/orders_and_users.db")
```

```python
query = sqlalchemy.text("SELECT * FROM orders")
orders = pd.read_sql(query, con=engine)
query = sqlalchemy.text("SELECT * FROM customers")
customers = pd.read_sql(query, con=engine)
```

```python
orders.head(10)
```

```python
customers.head(10)
```

```python
orders_and_customs_pd = orders.merge(right=customers,how = "left", left_on="CustomerID",right_on="CustomerID",sort=True)
orders_and_customs_pd
```

```python
statistic = orders.merge(right=customers,how = "inner", on="CustomerID").groupby(["Location", "Category"]).agg(orders_count = ("CustomerID","count")).reset_index()
statistic
```

```python
merged = orders.merge(customers, on="CustomerID", how="inner")
gaming_keyboards = merged[merged["Description"] == "Gaming Keyboard"]
best_location_pd = gaming_keyboards.groupby("Location").agg(total_qty=("Quantity", "sum")).sort_values("total_qty", ascending=False).head(1)
best_location_pd
```

```python
merged = orders.merge(customers, on="CustomerID", how="inner")
merged["cost"] = merged["Quantity"] * merged["UnitPrice"] * (1 - merged["Discount"])
high_income = merged[merged["Annual_Income"] > 2000]
high_income_categories = (
    high_income
    .groupby("Category")
    .agg(total_sum=("cost", "sum"))
    .sort_values("total_sum", ascending=False)
    .reset_index()
)

high_income_categories
```
