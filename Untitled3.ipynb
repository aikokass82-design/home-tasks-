```python
import pandas as pd
import sqlalchemy
from google.colab import drive
drive.mount('/content/drive')
```

```python
engine = sqlalchemy.create_engine("sqlite+pysqlite:////content/drive/MyDrive/Colab Notebooks/orders_and_users.db")
```

```python
query = sqlalchemy.sql.text("SELECT * FROM customers WHERE location = :City")
```

```python
orders_by_city = pd.read_sql(query, con=engine, params={"City": "Kyiv"})
orders_by_city
```

```python
query_3 = sqlalchemy.sql.text("SELECT * FROM customers ORDER BY Spending_score DESC LIMIT 3")
top_3 = pd.read_sql(query_3, con=engine)
top_3
```

```python
query_powerbank = """
SELECT SUM(quantity) as total_sold
FROM orders
WHERE Description = "Power Bank";
"""
df_powerbank = pd.read_sql(query_powerbank, con=engine)
print("\nВсего продано Power Bank:")
print(df_powerbank)
```

```python
query_credit = """
SELECT SUM(quantity) as total_paid
FROM orders
WHERE PaymentMethod LIKE "%card";
"""
result_df = pd.read_sql(query_credit, con=engine)
result_df
```
